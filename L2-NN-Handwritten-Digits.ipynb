```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Loading Dataset for training and testing   by using this command = keras.dataset.mnist.load_data()
# Loading X - train and Y - test dataset variables
(X_train, Y_train) , (X_test, Y_test) = keras.datasets.mnist.load_data()
```

```python
# Scaling the image from unint to float for getting better accuracy 
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_train_flattened.shape
```

```python
X_train[0]
```

```python
X_test.shape
```

```python
X_test_flattened = X_test.reshape(len(X_test), 28*28)
X_test_flattened.shape
```

```python
X_train_flattened[0]
```

```python
# Single Neural Network -- Training Dataset
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')  
])

model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train_flattened, Y_train, epochs= 20)
```

```python
# Testing Dataset on a tained model

model.evaluate(X_test_flattened, Y_test)
 
```

```python
model.predict(X_test_flattened)
```

```python
plt.matshow(X_test[2])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[2]
```

```python
# To find the maximum value in the y_predicted array index
np.argmax(y_predicted[2])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
Y_test[:5]
```

```python
# Build Confusin Metric Function 
cm = tf.math.confusion_matrix(labels=Y_test, predictions=y_predicted_labels)
cm 
```

```python
# Visualizing the confusion metrics
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Adding Hidden Layer in the Model

model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,),activation='relu'),
    keras.layers.Dense(100, activation='sigmoid')

])

model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train_flattened, Y_train, epochs= 20)
```

```python
model.evaluate(X_test_flattened, Y_test)
```

```python
 # Visualizing the confusion metrics
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
#Flatten  
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(100, activation='softmax')

])

model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train, Y_train, epochs= 5)
```
