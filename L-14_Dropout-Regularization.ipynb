```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 

warnings.filterwarnings('ignore')

df = pd.read_csv("./sonar_dataset.csv", header=None)
df.sample(5)
```

```python
df.shape
```

```python
# Checking if any of the column or row containing any null dataset

df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis='columns')
y = df[60]


y
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0

```

```python
y.value_counts()
```

```python
# Dataset dividing 80% training and 20% testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

```

```python
X_train.shape, X_test.shape
```

```python
# Model libraries
import tensorflow as tf
from tensorflow import keras

```

```python
# Building Artifical Neural Network
# Binary classification problem
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), # 60 columns means 60 neurons and 60 hidden layers
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid') # The last output layer
    
    
    
])

model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8) # Batch_size = it means 8 samples in each iterations
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

#round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
# adding Dropout layer

model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'), # 60 columns means 60 neurons and 60 hidden layers
    keras.layers.Dropout(0.5), # 50% neuron from hidden layer will dropout
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dropout(0.5), # Best practice is to add dropout layer after hidden layer
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid') # The last output layer
    
    
    
])

model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

model.fit(X_train, y_train, epochs=110, batch_size=8) # Batch_size = it means 8 samples in each iterations
```

```python
# Using dropout will drop some accuracy but increase the testing accuracy
model.evaluate(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report


y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred) # because using sigmoid function
print(classification_report(y_test, y_pred))
```
